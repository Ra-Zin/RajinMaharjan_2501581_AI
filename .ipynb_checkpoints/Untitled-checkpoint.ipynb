```python
i = [1,2,7,6,5]
highest_num = i[0]
for j in i:
    if j > highest_num:
        highest_num = j
print(highest_num)
```

```python
fact = 7
for i in range(fact-1,0, -1):
    fact = fact * i
print(fact) 
```

```python
for i in range(2, 100):
    is_prime = True
    for j in range (2,i):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i, end = " ")
```
